```python
import pandas as pd
import numpy as np
```

```python
data = {
    "comedy-score": [2, 9, 3, 8, 8, 9, 4, 2, 1],
    "action-score": [8, 3, 7, 2, 3, 5, 9, 9, 9]
}
```

```python
df = pd.DataFrame(data)
```

```python
x = df[["comedy-score", "action-score"]]
```

```python
from sklearn.cluster import KMeans
KM = KMeans(n_clusters=2)
KM.fit(x)
```

```python
KM.predict([[8, 3]])
```

```python
import matplotlib.pyplot as plt
```

```python
df["cluster"] = KM.labels_
```

```python
centroids = KM.cluster_centers_
plt.scatter(
    df["comedy-score"],
    df["action-score"],
    c=df["cluster"]
  )
plt.scatter(
    centroids[:, 0],
    centroids[:, 1],
    marker = "x",
    s = 100,
    linewidth = 5
)
```
