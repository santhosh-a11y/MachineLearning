```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("/content/student_depression_dataset.csv")
```

```python
x = df[["Academic Pressure"]]
y = df["Depression"]
```

```python
LR = LinearRegression()
LR.fit(x, y)
```

```python
a = float(input("what is your CGPA: "))
A = LR.predict([[a]])
if A >= 0.5:
  print("You are depressed")
else:
  print("You are not depressed")
```

```python
x_plot = x[::500]
y_plot = y[::500]
y_pred = LR.predict(x)
y_pred_plot = y_pred[::500]
plt.scatter(x_plot, y_plot)
plt.plot(x_plot, y_pred_plot, color='red')
```
