```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/user_behavior_dataset (1).csv")
```

```python
new_df = df[["Age",
             "Screen On Time (hours/day)"]]
```

```python
new_df.isnull()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
LR = LinearRegression()
```

```python
inp = new_df[["Age"]]
out = new_df["Screen On Time (hours/day)"]
```

```python
LR.fit(inp, out)
```

```python
LR.predict([[25]])
```

```python
y_pred = LR.predict(inp)
inp_plot = inp[::100]
out_plot = out[::100]
y_pred_plot = y_pred[::100]
```

```python
plt.scatter(inp_plot, out_plot)
plt.plot(inp_plot, y_pred_plot)
```
