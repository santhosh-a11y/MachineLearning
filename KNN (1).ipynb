```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
```

```python
data = {
    "study_hours": [1, 2, 3, 4, 5, 6, 7],
    "grade": ["c", "c", "b", "b", "b", "a", "a"]
}
```

```python
df = pd.DataFrame(data)
```

```python
grade_map = {"a": 1, "b": 2, "c": 3}
df["grade_encode"] = df["grade"].map(grade_map)
```

```python
print(df)
```

```python
x = df[["study_hours"]]
y = df["grade_encode"]
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x, y)
```

```python
knn.predict([[5]])
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_line = np.linspace(x.min(), x.max(), 300).reshape(-1, 1)
y_line = knn.predict(x_line)

plt.scatter(x, y)
plt.plot(x_line, y_line)

plt.yticks([1, 2, 3], ["A", "B", "C"])
```
